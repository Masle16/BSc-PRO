# Plot of Results

```python
# Setup stuff
import numpy as np
import matplotlib.pyplot as plt
```

## Pixel feature on full images

```python
# Load data
x = ['32x32', '64x64', '128x128']
norm = [63.1, 64.6, 65.0]
smooth = [65.7, 66.3, 65.7]
filt = [44.9, 45.7, 47.7]
smooth_filt = [51.3, 49.4, 50.6]

# Plot data
plt.xlabel('Image size [pixels]')
plt.ylabel('Accuracy [%]')
plt.plot(x, norm, label='Normal')
plt.plot(x, smooth, label='Smoothed')
plt.plot(x, filt, label='Filtered')
plt.plot(x, smooth_filt, label='Smoothed + filtered')
plt.legend()
plt.show()
```
